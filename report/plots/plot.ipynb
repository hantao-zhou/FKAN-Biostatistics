```python
import json
import matplotlib.pyplot as plt

# load json
with open("all_results.json", "r") as file:
    data = json.load(file)


models = ["REAL_NET", "EFF_NET", "ResNet18"]


plt.figure(figsize=(10, 6))

# extract model data and plot
for model in models:
    val_f1s = data[model]["val_f1s"]
    plt.plot(range(len(val_f1s)), val_f1s, label=f"{model}")

plt.xlabel("Epochs")
plt.ylabel("F1-Score")
plt.title("Validation F1-Scores for Different Models")
plt.legend()
plt.show()
```
